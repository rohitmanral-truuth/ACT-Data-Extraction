# Importing essential libraries

```python
import selenium
import time
from selenium import webdriver

from bs4 import BeautifulSoup
import pandas as pd
```

# Loading the `Webdriver`

```python
driver = webdriver.Chrome()
```

# Getting the access of `ACT Builder's` official website

```python
driver.get('https://www.accesscanberra.act.gov.au/s/public-registers/construction-licences?registerid=m0I4a00000000CHEAY')
time.sleep(2)
```

# Finding number of pages of the data 

```python
pages=driver.find_element_by_xpath('/html/body/div[4]/div[3]/div/div/div/div/div/div/div[3]/div/div/div[2]/c-cxs-public-registers-construction/div/div[6]/p').text
pages
```

```python
numbers = []
for word in pages.split():
    if word.isdigit():
        numbers.append(int(word))
print(numbers[1])
```

# Main program to get `Headers` & `all the data rows` of just the `first page`

```python
page_source = driver.page_source
soup =  BeautifulSoup(page_source, 'lxml')
            
headers = [i.text for i in soup.findAll('th')]
headers = headers[:-1]
print(headers)




data = [j.text for j in soup.findAll('td')]
    
all_names= data[::6]
    
data=data[1:]
all_license_no= data[::6]
    
data=data[1:]
all_exp= data[::6]
    
data=data[1:]
all_occupation= data[::6]
    
data=data[1:]
all_class= data[::6]
```

# Main program to get `all the data rows` of all the pages `till 460`

```python
for i in range(1, numbers[1]):
    
    driver.find_element_by_xpath('/html/body/div[4]/div[3]/div/div/div/div/div/div/div[3]/div/div/div[2]/c-cxs-public-registers-construction/div/div[6]/div[4]/button').click()
    
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    data = [j.text for j in soup.findAll('td')]
    
    name= data[::6]
    #name
    all_names= all_names+ name
    
    data=data[1:]
    lic_no= data[::6]
    #lic_no
    all_license_no=all_license_no+ lic_no
    
    data=data[1:]
    exp= data[::6]
    #exp
    all_exp=all_exp+exp
    
    data=data[1:]
    occupation= data[::6]
    #occupation
    all_occupation=all_occupation+occupation
    
    data=data[1:]
    class_build= data[::6]
    #class_build
    all_class=all_class+class_build
```

# Creating the final dataframe by zipping all the attribute's lists

```python
df = pd.DataFrame(list(zip(all_names, all_license_no, all_exp, all_occupation, all_class)),
               columns =['Name', 'License Number','Expiry date','Occupation','Class'])
df
```
